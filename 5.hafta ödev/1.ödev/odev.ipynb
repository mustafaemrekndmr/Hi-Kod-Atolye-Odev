```python
Notlar = {}

def ogrenciBilgiGirisi():
    isim = input ("Öğrencinin İsmini Giriniz.")
    mat = int(input("Öğrencinin Matematik Notunu Giriniz."))
    fiz = int(input("Öğrencinin Fizik Notunu Giriniz."))
    kim = int(input("Öğrencinin Kimya Notunu Giriniz."))
    
    Notlar[isim] = {"Matematik": mat, "Fizik": fiz, "Kimya": kim}
    print("İşlem Başarılı!")
    
def notGoster():
    isim = input("Hangi Öğrencinin Notu Güncellenecek?")
    if isim in Notlar:
            print(Notlar[isim])
    else:
        print(f"{isim} adlı öğrenci bulunamadı.")
        
def notGuncelleme():
    isim = input("Notunu güncellemek istediğiniz öğrencinin ismi nedir? ")
    if isim in Notlar:
        ders = input(f"Hangi dersin notunu güncellemek istiyorsunuz?")
        if ders in Notlar[isim]:
            yeni_not = int(input(f"Yeni {ders} notunu girin: "))
            Notlar[isim][ders] = yeni_not
            print(f"{isim} için {ders} notu başarıyla güncellendi.")
        else:
            print(f"{ders} dersi bulunamadı.")
    else:
        print(f"{isim} adlı öğrenci bulunamadı.")
        
def tumNotlar():
    print("\nTüm Öğrenciler ve Notları:")
    for isim, notlar in Notlar.items():
        print(f"{isim}: Matematik: {notlar['Matematik']}, Fizik: {notlar['Fizik']}, Kimya: {notlar['Kimya']}")

while True:
    print("\n-- Öğrenci Not Sistemi --")
    print("1. Öğrenci Ekle")
    print("2. Öğrenci Notu Göster")
    print("3. Öğrenci Notu Güncelle")
    print("4. Tüm Öğrenciler ve Notları")
    print("5.Çıkış")
    
    secim = input("Bir işlem seçin (1-5):")
    
    if secim == '1':
        ogrenciBilgiGirisi()
    elif secim == '2':
        notGoster()
    elif secim == '3':
        notGuncelleme()
    elif secim == '4':
          tumNotlar()
    elif secim == '5':
        print("Programdan çıkılıyor...")
        break
    else:
        print("Geçersiz seçim. Lütfen 1-5 arasında bir değer giriniz.")
```
